# Assignment P1: Clustering

In this assignment, we will load the in-built `USArrests` to perform clustering

### Loading the `USArrests` dataset into variable `df`

```r
df<-USArrests
```

```r
head(USArrests)
```

Now, we will perform cluster analysis using Hierarchical clustering to determine the value of k(number of clusters) for K-Means clustering 

### Distance Matrix Calculation

```r
state_dm<-dist(df)
```

###  Hierarchical Clustering

```r
hir_clust<-hclust(state_dm,method = "complete")
```

```r
hir_clust
```

```r
plot(hir_clust,labels = rownames(df), ylab = "Distance")
```

```r
plot(hir_clust,labels = rownames(df), ylab = "Distance")
abline(h=200, col="red", lwd=2) 
abline(h=150, col="blue", lwd=2)
rect.hclust(hir_clust, k=3)
```

```r
set.seed(1234)
library(cluster)
kmeans_clust<-kmeans(df,centers = 3, nstart = 20)
```

```r
kmeans_clust$centers
```

```r
df$cluster<-kmeans_clust$cluster
```
